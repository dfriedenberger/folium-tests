```python
import folium

# http://python-visualization.github.io/folium/quickstart.html#Getting-Started

m = folium.Map(location=[52.52437, 13.41053],
           zoom_start=12,
           tiles='https://tiles.autobahn.de/osm_tiles/{z}/{x}/{y}.png',
           attr='autobahn.de')
#m = folium.Map(location=[52.52437, 13.41053])
```

```python
# https://gist.github.com/LilithWittmann/06bd153317b635e7b622651f5cfd95ea
import json
import urllib.request as urllib2
url = "https://verkehr.autobahn.de/o/autobahn/A3/services/webcam"
response = urllib2.urlopen(url)
data = response.read()
values = json.loads(data)    

for webcam in values['webcam']:
    #print(webcam['coordinate'],webcam['title'])
    lat = webcam['coordinate']['lat']
    lon = webcam['coordinate']['long']
    folium.Marker(
        [lat , lon], popup='<img src="{}">'.format(webcam['imageurl']), tooltip=webcam['title']
    ).add_to(m)
    
```
